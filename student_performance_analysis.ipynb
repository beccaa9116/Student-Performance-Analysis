### Average differences between top and worst students

```python
# exam score difference between top and worst students
exam_score_diff = top_students[['Exam_Score']].mean() - worst_students[['Exam_Score']].mean()
print("Top students have on average", round(float(exam_score_diff), 2), "points more than worst students.")

```

```python
# previous scores difference between top and worst students
previous_scores_diff = top_students[['Previous_Scores']].mean() - worst_students[['Previous_Scores']].mean()
print("Top students have scored", round(float(previous_scores_diff), 2), "points more on the previous exam than worst students.")


```

```python
#tutoring sessions difference between top and worst students
tutoring_sessions_diff = top_students[['Tutoring_Sessions']].mean() - worst_students[['Tutoring_Sessions']].mean()
print("Top students have attended", round(float(tutoring_sessions_diff), 2), "more tutoring sessions than worst students.")

```

```python
#studied more hours
hours_studied_diff = top_students[['Hours_Studied']].mean() - worst_students[['Hours_Studied']].mean()
print("Top students have studied", round(float(hours_studied_diff), 2), "hours more than worst students.")


```

### Comparing the best performing student with the worst performing student

```python
#compare the best performing student with the worst performing student
best_student = data.loc[data['Exam_Score'].idxmax()]
worst_student = data.loc[data['Exam_Score'].idxmin()]
```

```python
best_student
```

```python
worst_student
```

```python
print("The best performing student has an exam score of", best_student['Exam_Score'], "and the worst performing student has an exam score of", worst_student['Exam_Score'])
```

```python
#the best performing student has studied on average x hours more than the worst performing student
hours_studied_diff = best_student['Hours_Studied'] - worst_student['Hours_Studied']
print("The best performing student has studied", round(float(hours_studied_diff), 2), "hours more than the worst performing student.")
```

```python
#difference in previous scores
previous_scores_diff = best_student['Previous_Scores'] - worst_student['Previous_Scores']
print("The best performing student has scored", round(float(previous_scores_diff), 2), "points more on the previous exam than the worst performing student.")
```

```python
#difference in tutoring sessions
tutoring_sessions_diff = best_student['Tutoring_Sessions'] - worst_student['Tutoring_Sessions']
print("The best performing student has attended", round(float(tutoring_sessions_diff), 2), "more tutoring sessions than the worst performing student.")
```
